## Impulse Response

The impulse (or delta) function $\delta(t)$ 



```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#listing 4.1 specdelta.m
# modified for exercise 4.5
time = 2
Ts = 1.0/100
t = np.linspace(0, time, int(time/Ts) + 1)
x = np.zeros(len(t))
spacing = 20
for i in range(len(t)):
    if i % spacing == 1:
        x[i] = 1

fig, axs = plt.subplots(2, 1)
axs[0].plot(t, x)

fourier = np.fft.fft(x)
n = len(x)
freq = np.fft.fftfreq(n, d=Ts)
axs[1].plot(freq, abs(fourier))
## answer to part c
print((1/Ts)/spacing)
```

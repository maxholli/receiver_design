```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
## p. 42
f=10     #freq
time=2   #length of time
Ts = 1/1000.0  # time interval
t = np.linspace(Ts, time, time/Ts)
x = np.cos(2*np.pi*f*t)
```

```python
plt.plot(t,x)
plt.show()
```

```python
fourier = np.fft.fft(x)
n = len(x)
freq = np.fft.fftfreq(n, d=Ts)

plt.plot(freq, abs(fourier))
plt.show()
```

```python
#p. 43 an exponential pulse e^(-t)
time = 10   #length of time
Ts = 1/10000.0  # time interval
t = np.linspace(Ts, time, time/Ts)
x = np.exp(-1*t)
plt.plot(t,x)
plt.show()
```

```python
fourier = np.fft.fft(x)
n = len(x)
freq = np.fft.fftfreq(n, d=Ts)

plt.plot(freq, abs(fourier))
plt.show()
```

```python
#p. 43 an gaussian pulse e^(-t^2)
time = 4   #length of time
Ts = 1/5000.0  # time interval
t = np.linspace(Ts-2, time-2, time/Ts)
x = np.exp(-1*(t*t))
plt.plot(t,x)
plt.show()
```

```python
fourier = np.fft.fft(x)
n = len(x)
freq = np.fft.fftfreq(n, d=Ts)

plt.plot(freq, abs(fourier))
plt.show()
```
